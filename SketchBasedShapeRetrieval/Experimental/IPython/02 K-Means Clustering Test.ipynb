```python
import numpy
import math
import scipy
import scipy.io
import scipy.misc
import matplotlib
import matplotlib.pyplot
import matplotlib.cm
import os
import IPython
import IPython.parallel
import itertools
import random
import sklearn
import sklearn.cluster
```

```python
local_features_mat = scipy.io.loadmat("local_feature_corpus_reduced.pca.2d.mat")
local_features     = local_features_mat["local_features_reduced_pca_2d"]
```

```python
k    = 1000
seed = 0

k_means_clustering = sklearn.cluster.KMeans(k=k, random_state=seed, verbose=True)
k_means_clustering.fit(local_features)
```

```python
local_feature_cluster_assignments = k_means_clustering.labels_
cluster_centers                   = k_means_clustering.cluster_centers_

figsize(10,10)
axis("equal")

color_indices = linspace(0, 1, k)
scatter(local_features[:,0], local_features[:,1], marker='o', c=color_indices[local_feature_cluster_assignments], s=40, alpha=0.15);
scatter(cluster_centers[:,0], cluster_centers[:,1], marker='o', c=color_indices, s=40);
```

```python
k    = 1000
seed = 0

minibatch_k_means_clustering = sklearn.cluster.MiniBatchKMeans(k=k, random_state=seed, verbose=True)
minibatch_k_means_clustering.fit(local_features)
```

```python
local_feature_cluster_assignments = minibatch_k_means_clustering.labels_
cluster_centers                   = minibatch_k_means_clustering.cluster_centers_

figsize(10,10)
axis("equal")

color_indices = linspace(0, 1, k)
scatter(local_features[:,0], local_features[:,1], marker='o', c=color_indices[local_feature_cluster_assignments], s=40, alpha=0.15);
scatter(cluster_centers[:,0], cluster_centers[:,1], marker='o', c=color_indices, s=40);
```
